```python
C = np.array([[4,1],[2,3]])
eval, evec = np.linalg.eig(C) # columns are eigenvectors

print(f'Eigenvalues: {eval}')
print(f'Eigenvectors normalized: \n{evec.tolist()}')

v1_norm = evec[:, 0]  # ~ [0.7071, 0.7071]
v2_norm = evec[:, 1]  # ~ [-0.4472, 0.8944]

v1 = v1_norm / v1_norm[0]
v2 = v2_norm / v2_norm[1]

print(f'''\nEigenvectors neat: 
{np.column_stack([v1,v2])}''')
```

## Problem 1

```python
import numpy as np
from itertools import combinations

def find_bases(vectors):
    # Convert to numpy arrays
    arrs = [np.array(v) for v in vectors]
    
    # Check all vectors have same dimension
    dims = [len(v) for v in arrs]
    if len(set(dims)) != 1:
        return "The list of vectors contains vectors of different dimensions."
    
    D = dims[0]
    
    # Stack into matrix
    M = np.vstack(arrs)
    rank = np.linalg.matrix_rank(M)
    
    if rank < D:
        return "There are less than D independent vectors in the set."
    
    # Otherwise, find all bases (all subsets of size D with full rank)
    bases = []
    for combo in combinations(arrs, D):
        mat = np.vstack(combo)
        if np.linalg.matrix_rank(mat) == D:
            bases.append(combo)
    
    return bases

# Example
L = [[1,2,3,4],[5,6,7,8],[6,8,10,12],[4,4,4,4],[3,2,1,0]]
print(find_bases(L))
```

## Problem 2

```python
def projection(v, w):
    v = np.array(v, dtype=float)
    w = np.array(w, dtype=float)
    
    if np.allclose(w, 0):
        raise ValueError("Cannot project onto the zero vector.")
    
    proj = (np.dot(v, w) / np.dot(w, w)) * w
    return proj.astype(int) if np.allclose(proj, np.round(proj)) else proj

# Example
v = [1,2,3,4]
w = [0,0,1,0]
print(projection(v, w))
```
