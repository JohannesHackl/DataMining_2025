```python
# add new names, 
# remove existing ones
# display all names in the list
# save the file to a file
# load the file when the program starts.
```

```python
students = []

check = False
action = 0

while True:
    while check == False:
        action = input("""
        Choose an option:
            1. Add a Student
            2. Remove a Student
            3. Display all Students
            4. Exit
        """)
        
        action = int(action)

        if action not in [1,2,3,4]: # != 1 and action != 2 and action != 3 and action != 4:
            print("not valid, try again.") 
        else:
            check = True

    if action == 1:
        name = input("Enter the students' name to add.")
        students.append(name)
    elif action == 2:
        name = input("Enter the students' name to remove.")
        students.remove(name)
    elif action == 3:
        for i in students:
            print(f"{i}, ")
    elif action == 4:
        print("Good Bye")
        break

```
